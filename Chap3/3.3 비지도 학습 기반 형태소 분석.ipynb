```python
from soynlp.word import WordExtractor

corpus_fname = 'data\processed\processed_ratings.txt'
model_fname = 'data\processed\soyword.model'
```

```python
sentence = [sent.strip() for sent in open(corpus_fname, 'r', encoding='utf-8').readlines()]
sentence
```

```python
word_extractor = WordExtractor(min_frequency=100,
                              min_cohesion_forward=0.05, # 응집확률
                              min_right_branching_entropy=0.0) # 단어 앞뒤로 다양한 조사어미단어 등장하는 빈도
word_extractor.train(sentence)
word_extractor.save(model_fname)
```

```python
import math
from soynlp.tokenizer import LTokenizer

word_extractor = WordExtractor(min_frequency=100,
                              min_cohesion_forward=0.05, # 응집확률
                              min_right_branching_entropy=0.0) # 단어 앞뒤로 다양한 조사어미단어 등장하는 빈도
word_extractor.load(model_fname)
scores = word_extractor.word_scores()
```

```python
scores
```

```python
scores.keys()
```

```python
scores = {key:(scores[key].cohesion_forward * math.exp(scores[key].right_branching_entropy)) for key in scores.keys()}
scores
```

```python
tokenizer = LTokenizer(scores=scores)
```

```python
tokens = tokenizer.tokenize('아버지가방에들어가신다')
tokens
```
